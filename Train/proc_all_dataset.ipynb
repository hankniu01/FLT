```python
import os
import json
```

```python
datasets = ['Restaurants', 'Laptop', 'Tweets']
```

```python
fns = [f"Test.json", f"Train.json"]
```

```python
data_dir = r"/nfsfile/niuhao/project/aspect_process_for_thesis/FLT/Dataset"
```

```python
alldataset = {}
for split, name in zip(["test", "train"], fns):
    alldataset[split] = []
    for dataset in datasets:
        paths = os.path.join(data_dir, dataset)
        fp = os.path.join(paths, name)
        with open(fp, "r", encoding="utf-8") as f:
            data = json.load(f)
        
        alldataset[split] += data
        
    json.dump(alldataset[split], open(os.path.join(data_dir, 'alldataset', name), 'w', encoding="utf-8"))
```
